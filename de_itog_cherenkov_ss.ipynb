# Загрузка внешних данных в таблицы на сервере Oracle с последующим обновлением витрин


```python
import pandas as pd
import sqlalchemy as adb
from sqlalchemy import MetaData
import cx_Oracle as ora
import datetime as dt

# импорт данных
l_orders  =         pd.read_excel(r'D:\DE\REBOOT_DE2\dataset_800.xlsx', sheet_name = 'orders', header = 0)
l_order_payments =  pd.read_excel(r'D:\DE\REBOOT_DE2\dataset_800.xlsx', sheet_name = 'order_payments', header = 0)  
l_customers =       pd.read_excel(r'D:\DE\REBOOT_DE2\dataset_800.xlsx', sheet_name = 'customers', header = 0)
l_order_items =     pd.read_excel(r'D:\DE\REBOOT_DE2\dataset_800.xlsx', sheet_name = 'order_items', header = 0)
l_products =        pd.read_excel(r'D:\DE\REBOOT_DE2\dataset_800.xlsx', sheet_name = 'products', header = 0)
l_sellers =         pd.read_excel(r'D:\DE\REBOOT_DE2\dataset_800.xlsx', sheet_name = 'sellers', header = 0)

# подключение к oracle
l_user = 'cherenkov_ss'
l_pass = open(r'D:\DE\REBOOT_DE2\pass.txt', 'r').read()
l_tns = ora.makedsn('13.95.167.129', 1521, service_name = 'pdb1')

l_conn_ora = adb.create_engine(r'oracle://{p_user}:{p_pass}@{p_tns}'.format(
    p_user = l_user
    , p_pass = l_pass
    , p_tns = l_tns
    )
    )
l_meta = MetaData(l_conn_ora)
l_meta.reflect()


print('Начался процесс загрузки данных и обновления витрин')
# Вставка данных в ora в таблицу customers_csv
l_customers_excel = l_meta.tables['customers_excel']
l_customers_list = l_customers.values.tolist()

print('\nНачалась вставка данных в таблицу customers_excel')
for i in l_customers_list:
    l_customers_excel.insert([l_customers_excel.c.customer_id
                            , l_customers_excel.c.customer_unique_id
                            , l_customers_excel.c.customer_zip_code_prefix
                            , l_customers_excel.c.customer_city
                            , l_customers_excel.c.customer_state]
                           ).values(
                               customer_id =                i[0]
                               , customer_unique_id =       i[1]
                               , customer_zip_code_prefix = i[2]
                               , customer_city =            i[3]
                               , customer_state =           i[4]
                           ).execute()
print('Данные загружены в таблицу customers_excel')




# Вставка данных в ora в таблицу order_items_excel
l_order_items_excel = l_meta.tables['order_items_excel']
l_order_items_list = l_order_items.values.tolist()

print('\nНачалась вставка данных в таблицу order_items_excel')
for i in l_order_items_list:
    l_order_items_excel.insert([l_order_items_excel.c.order_id
                            , l_order_items_excel.c.order_item_id
                            , l_order_items_excel.c.product_id
                            , l_order_items_excel.c.seller_id
                            , l_order_items_excel.c.shipping_limit_date
                            , l_order_items_excel.c.price
                            , l_order_items_excel.c.freight_value]
                           ).values(
                               order_id =               i[0]
                               , order_item_id =        i[1]
                               , product_id =           i[2]
                               , seller_id =            i[3]
                               , shipping_limit_date =  i[4]
                               , price =                i[5]
                               , freight_value =        i[6]
                           ).execute()
                          
print('Данные загружены в таблицу order_items_excel')




# Вставка данных в ora в таблицу products_excel
l_products_excel = l_meta.tables['products_excel']
l_products_list = l_products.values.tolist()

print('\nНачалась вставка данных в таблицу products_excel')
for i in l_products_list:
    l_products_excel.insert([l_products_excel.c.product_id
                            , l_products_excel.c.product_category_name]
                           ).values(
                               product_id =                i[0]
                               , product_category_name =   i[1]
                           ).execute()
print('Данные загружены в таблицу products_excel')




#Вставка данных в ora в таблицу sellers_excel
l_sellers_excel = l_meta.tables['sellers_excel']
l_sellers_list = l_sellers.values.tolist()

print('\nНачалась вставка данных в таблицу sellers_excel')
for i in l_sellers_list:
    l_sellers_excel.insert([l_sellers_excel.c.seller_id
                            , l_sellers_excel.c.seller_zip_code_prefix
                            , l_sellers_excel.c.seller_city
                            , l_sellers_excel.c.seller_state]
                           ).values(
                               seller_id =                 i[0]
                               , seller_zip_code_prefix =  i[1]
                               , seller_city =             i[2]
                               , seller_state =            i[3]
                           ).execute()
                          
print('Данные загружены в таблицу sellers_excel')



# Вставка данных в ora в таблицу order_payments_excel
l_order_payments_excel = l_meta.tables['order_payments_excel']
l_order_payments_list = l_order_payments.values.tolist()

print('\nНачалась вставка данных в таблицу order_payments_excel')
for i in l_order_payments_list:
    l_order_payments_excel.insert([l_order_payments_excel.c.order_id
                            , l_order_payments_excel.c.payment_sequential
                            , l_order_payments_excel.c.payment_type
                            , l_order_payments_excel.c.payment_installments
                            , l_order_payments_excel.c.payment_value]
                           ).values(
                               order_id =                i[0]
                               , payment_sequential =    i[1]
                               , payment_type =          i[2]
                               , payment_installments =  i[3]
                               , payment_value =         i[4]
                           ).execute()
                          
print('Данные загружены в таблицу order_payments_excel')




# Вставка данных в ora в таблицу orders_excel
l_orders_excel = l_meta.tables['orders_excel']
l_orders_list = l_orders.values.tolist()

print('\nНачалась вставка данных в таблицу orders_excel')
for i in l_orders_list:
    l_orders_excel.insert([l_orders_excel.c.order_id
                            , l_orders_excel.c.customer_id
                            , l_orders_excel.c.order_status
                            , l_orders_excel.c.order_purchase_timestamp
                            , l_orders_excel.c.order_approved_at
                            , l_orders_excel.c.order_delivered_carrier_date
                            , l_orders_excel.c.order_delivered_customer_date
                            , l_orders_excel.c.order_estimated_delivery_date]
                           ).values(
                               order_id =                           i[0]
                               , customer_id =                      i[1]
                               , order_status =                     i[2]
                               , order_purchase_timestamp =         i[3]
                               , order_approved_at =                i[4]
                               , order_delivered_carrier_date =     i[5]
                               , order_delivered_customer_date =    i[6]
                               , order_estimated_delivery_date =    i[7]
                           ).execute()
                          
print('Данные загружены в таблицу orders_excel')

print('\nВсе данные загружены.')

print('\nНачался процесс обновления витрин.')
def pervii_sloi():
    l_conn_ora.execute(adb.text('BEGIN CHERENKOV_SS_orders_pack.vitrina_pervii_sloi; END;'))
    print('Таблица с первым слоем обновлена')

def vitrina_agg_2zakaza_sum500():
    l_conn_ora.execute(adb.text('BEGIN CHERENKOV_SS_orders_pack.vitrina_agg_2zakaza_sum500; END;'))
    print('Обновлена витрина по клиентам с 2-мя и более заказов на сумму выше 500 у.е. ')    

def vitrina_agg_city_year():
    l_conn_ora.execute(adb.text('BEGIN CHERENKOV_SS_orders_pack.vitrina_agg_city_year; END;'))
    print('Обновлена витрина по городам и суммам заказов в разрезе лет') 

def vitrina_slice_month_orders():
    l_conn_ora.execute(adb.text('BEGIN CHERENKOV_SS_orders_pack.vitrina_slice_month_orders; END;'))
    print('Обновлена витрина по продажам в разрезе месяцев') 
    
    
pervii_sloi()
vitrina_agg_2zakaza_sum500()
vitrina_agg_city_year()
vitrina_slice_month_orders()
    
print('\nВсе витрины обновлены')
print('\nЗавершен процесс загрузки данных и обновления витрин')
```
